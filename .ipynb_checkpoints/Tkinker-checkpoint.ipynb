```python
import tkinter as tk
```

```python
root = tk.Tk()
```

```python
root.title("my first Tkinter window") 
```

```python
root.mainloop() #questo va runnato alla fine altrimenti il resto da errore
```

```python
welcome_label = tk.Label(root,text = "Welcome to tkinter")
```

```python
welcome_label.pack ()
```

```python
def on_button_click():
    welcome_label.config (text="button clicked")
    
```

```python
click_me_button = tk.Button(root, text="click me !", command=on_button_click)
```

```python
click_me_button.pack ()
```
