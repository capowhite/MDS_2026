```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import numpy as np

# Dati
x = np.arange(5)
y = np.zeros(5)
z = np.zeros(5)

dx = np.ones(5) * 0.5
dy = np.ones(5) * 0.5
dz = [5, 7, 3, 8, 6]

# Figura 3D
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

ax.bar3d(x, y, z, dx, dy, dz)

ax.set_xlabel("Days")
ax.set_ylabel("Y")
ax.set_zlabel("Rain mm")

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Matrice di dati
data = np.array([
    [1, 2, 3, 4],
    [4, 5, 6, 7],
    [7, 8, 9, 10]
])

plt.figure(figsize=(6,5))
plt.imshow(data, cmap="viridis")
plt.colorbar(label="Intensity")
plt.title("Rain Heatmap")
plt.xlabel("Days")
plt.ylabel("Cities")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(0)
x = np.arange(1, 31)
y = np.cumsum(np.random.randn(len(x))) + 20
y2 = y + np.random.randn(len(x))*0.8

mean = (y + y2) / 2
std = np.std(np.vstack([y, y2]), axis=0)

plt.figure(figsize=(9,4))
plt.plot(x, y, label="Serie A")
plt.plot(x, y2, label="Serie B")
plt.fill_between(x, mean-std, mean+std, alpha=0.2, label="1 std")

peak_i = np.argmax(mean)
plt.scatter(x[peak_i], mean[peak_i])
plt.annotate("Picco", (x[peak_i], mean[peak_i]), xytext=(x[peak_i]+2, mean[peak_i]+1),
             arrowprops=dict(arrowstyle="->"))

plt.title("Trend con banda di confidenza")
plt.xlabel("Giorni")
plt.ylabel("Valore")
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ["A", "B", "C", "D", "E"]
values = [12, 7, 15, 9, 13]

plt.figure(figsize=(7,4))
bars = plt.bar(labels, values)

for b in bars:
    h = b.get_height()
    plt.text(b.get_x() + b.get_width()/2, h + 0.3, f"{h}", ha="center", va="bottom")

plt.ylim(0, max(values) + 4)
plt.grid(axis="y", alpha=0.3)
plt.title("Bar chart con etichette")
plt.ylabel("Valore")
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

categories = ["Company A", "Company B", "Company C", "Company D"]
values = [20, 50, 10, 40]
colors = ["green", "blue", "red", "orange"]

fig = plt.figure()
ax = fig.add_subplot(111, projection="3d")

# Posizioni
x = np.arange(len(categories))
y = np.zeros(len(categories))      # tutte sullo stesso asse Y
z = np.zeros(len(categories))      # base delle barre

dx = np.ones(len(categories)) * 0.5
dy = np.ones(len(categories)) * 0.5
dz = values                        # altezza barre

ax.bar3d(x, y, z, dx, dy, dz, color=colors)

# Etichette asse X con nomi aziende
ax.set_xticks(x)
ax.set_xticklabels(categories)

ax.set_zlabel("Revenues")
ax.set_title("Sample Plot")

plt.show()
```
