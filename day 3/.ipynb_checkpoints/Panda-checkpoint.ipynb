```python
pivot = df_house.pivot_table(
        index = "rooms", 
        aggfunc = "count", 
        values = "city")
```

```python
pivot.plot(kind = "bar")
```

```python
df2 = df_house[df_house["total (R$)"]<10000]
```

```python
max_price = df2["total (R$)"].max()
print(f" The maximum price is {max_price}")
```

```python
threshold = max_price * 0.9
print (f"the maximum threshold is {threshold}")
```

```python
top_df= df2[df_house["total (R$)"]>threshold]

top_df
```

```python
top_df_sorted = top_df.sort_values( 
    by = "total (R$)",
    ascending = False )
top_df_sorted.head(10)
```
