```python
import numpy as np
import pandas as pd

def preprocess_data(df):
    df = df.copy()  # evita problemi di chained assignment

    # Drop colonne inutili (se esistono)
    df.drop(columns=["PassengerId", "Name", "Ticket", "Cabin"], inplace=True, errors="ignore")

    # Embarked: riempi i NaN e (se vuoi) poi puoi anche dropparla
    df["Embarked"] = df["Embarked"].fillna("S")

    # Se davvero vuoi eliminare Embarked:
    df.drop(columns=["Embarked"], inplace=True, errors="ignore")

    # Sex -> numerico
    df["Sex"] = df["Sex"].map({"male": 1, "female": 0})

    # Family features (attenzione al nome: familySize coerente)
    df["familySize"] = df["SibSp"] + df["Parch"]
    df["IsAlone"] = np.where(df["familySize"] == 0, 1, 0)  # usa lo stesso nome

    # FareBin: qcut si usa così
    df["FareBin"] = pd.qcut(df["Fare"], q=4, labels=False)

    # AgeBin: labels deve essere False o una lista di etichette
    df["AgeBin"] = pd.cut(df["Age"], bins=[0, 12, 20, 40, 60, np.inf], labels=False)

    return df
```

```python
def fill_missining_ages(df):
    age_fill_map = {}
    for pclass in df["Pclass"].unique():
        if pclass not in age_fill_map:
            age_fill_map[pclass] =df [df["Pclass"] == pclass] ["Age"].median()
    df["Age"] = df.apply(lambda row: age_fill_map[row["Pclass"]] if pd.isnull(row["Age"]) else row["Age"],axis=1)

    
```

```python
data = preprocess_data(data)
x = data.drop(columns = ["Survived"])
y = data ["Survived"]
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.25,random_state=42)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier
from sklearn.preprocessing import MinMaxScaler
from sklearn.impute import SimpleImputer

# 1) Imputer (prima dello scaler!)
imputer = SimpleImputer(strategy="median")
x_train = imputer.fit_transform(x_train)
x_test  = imputer.transform(x_test)

# 2) Scaling
scaler = MinMaxScaler()
x_train = scaler.fit_transform(x_train)
x_test  = scaler.transform(x_test)

def tune_model(x_train, y_train):
    param_grid = {
        "n_neighbors": range(1, 21),
        "metric": ["euclidean", "manhattan", "minkowski"],  # FIX
        "weights": ["uniform", "distance"]
    }

    model = KNeighborsClassifier()
    grid_search = GridSearchCV(model, param_grid, cv=5, n_jobs=-1)
    grid_search.fit(x_train, y_train)
    return grid_search.best_estimator_

best_model = tune_model(x_train, y_train)
```

```python
# --- IMPORTS ---
import numpy as np
import pandas as pd

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# --- FUNZIONE DI VALUTAZIONE ---
def evaluate_model(model, x_test, y_test):
    predictions = model.predict(x_test)
    acc = accuracy_score(y_test, predictions)
    matrix = confusion_matrix(y_test, predictions)
    return acc, matrix, predictions

# --- ESECUZIONE (assume che best_model, x_test, y_test esistano già) ---
accuracy, matrix, predictions = evaluate_model(best_model, x_test, y_test)

print(f"Accuracy: {accuracy * 100:.2f}%")
print("Confusion Matrix:")
print(matrix)

# (opzionale) report completo
print("\nClassification Report:")
print(classification_report(y_test, predictions))
```

```python
# --- IMPORTS ---
import matplotlib.pyplot as plt
import seaborn as sns

# --- FUNZIONE PER PLOT CONFUSION MATRIX ---
def plot_model(matrix):
    plt.figure(figsize=(10, 7))
    
    sns.heatmap(
        matrix,
        annot=True,
        fmt="d",
        cmap="Blues",
        xticklabels=["Not Survived", "Survived"],
        yticklabels=["Not Survived", "Survived"]
    )
    
    plt.title("Confusion Matrix")
    plt.xlabel("Predicted Values")
    plt.ylabel("True Values")
    
    plt.show()

# --- CHIAMATA FUNZIONE ---
plot_model(matrix)
```
